```python
# 3. 라이브러리 임포트
import os
import librosa
import soundfile as sf
```

```python
import soundfile as sf

# 4. 설정 변수 정의 (Google 드라이브 경로 사용)
input_folder = '' # 원본 wav 파일이 저장된 폴더 경로
output_folder = '' # 5초 자른 파일을 저장할 폴더 경로
target_duration = 5.0 # 타겟 길이(초)

# 5. 파일 처리
num_saved = 0
num_skipped = 0

for animalname in os.listdir(input_folder):
    index = 0

    for filename in os.listdir(os.path.join(input_folder, animalname)):
        file_path = os.path.join(input_folder, animalname, filename)

        # 오디오 파일 로드
        audio, sr = librosa.load(file_path, sr=None)

        # 오디오 파일의 길이를 초 단위로 계산
        duration = librosa.get_duration(y=audio, sr=sr)

        # 파일이 5초 이상인 경우 앞부분 5초만 추출하여 저장
        if duration >= target_duration:
            index += 1
            num_saved += 1

            # 앞의 5초 추출
            clipped_audio = audio[:int(sr * target_duration)]

            # 새로운 파일 경로 설정
            output_path = os.path.join(output_folder, f"{animalname}_{index}.wav")

            # 파일 저장
            sf.write(output_path, clipped_audio, sr)
            print(f"++Saved : {filename}({round(duration,3)} seconds) as {animalname}_{index}.wav")
        else:
            num_skipped += 1
            print(f"--Skipped : {filename}({round(duration,3)} seconds)")

print ("\n=======================================================================================")
print(f"Total files saved: {num_saved}")
print(f"Total files skipped: {num_skipped}")
```
